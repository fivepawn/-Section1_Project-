```python
# 판매량 단위 통합 

def convert_sales_to_float(sales_value):
    if 'K' in sales_value:
        return float(sales_value.replace('K', '')) / 1000
    elif 'M' in sales_value:
        return float(sales_value.replace('M', ''))
    else:
        return float(sales_value)
```

```python
df['NA_Sales'] = df['NA_Sales'].apply(convert_sales_to_float)
df['EU_Sales'] = df['EU_Sales'].apply(convert_sales_to_float)
df['JP_Sales'] = df['JP_Sales'].apply(convert_sales_to_float)
df['Other_Sales'] = df['Other_Sales'].apply(convert_sales_to_float)
```

```python
# "Name" 열의 값 중 중복을 허용한 값들을 추출하여 "duplicated_names" 변수에 저장합니다.
duplicated_names = df[df.duplicated(subset=['Name'], keep=False)]['Name'].unique()

# 같은 이름을 가진 게임 중에서 Genre, Publisher, Year가 NaN값인 경우, 이를 채웁니다.
for name in duplicated_names:
    # 해당 이름을 가진 게임 중에서 Genre, Publisher, Year가 NaN값이 아닌 경우, 이 값을 "genre", "publisher", "year" 변수에 저장합니다.
    genre = df[(df['Name'] == name) & df['Genre'].notnull()]['Genre'].values
    publisher = df[(df['Name'] == name) & df['Publisher'].notnull()]['Publisher'].values
    year = df[(df['Name'] == name) & df['Year'].notnull()]['Year'].values
    
    # Genre, Publisher, Year가 NaN값인 데이터를 추출합니다.
    nan_genre = df[(df['Name'] == name) & df['Genre'].isnull()]
    nan_publisher = df[(df['Name'] == name) & df['Publisher'].isnull()]
    nan_year = df[(df['Name'] == name) & df['Year'].isnull()]
    
    # Genre, Publisher, Year가 NaN값인 데이터를 채웁니다.
    if len(genre) > 0:
        df.loc[nan_genre.index, 'Genre'] = genre[0]
    if len(publisher) > 0:
        df.loc[nan_publisher.index, 'Publisher'] = publisher[0]
    if len(year) > 0:
        df.loc[nan_year.index, 'Year'] = year[0]
```

```python
# 지역은 어떻게 구분했나요?

grouped = df.groupby('Genre')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()

# 각 지역에서 가장 많이 판매된 게임 장르 구하기
max_genre_na = df.groupby('Genre')['NA_Sales'].sum().nlargest(5).index
max_genre_eu = df.groupby('Genre')['EU_Sales'].sum().nlargest(5).index
max_genre_jp = df.groupby('Genre')['JP_Sales'].sum().nlargest(5).index
max_genre_other = df.groupby('Genre')['Other_Sales'].sum().nlargest(5).index

print('North America:', max_genre_na)
print('Europe:', max_genre_eu)
print('Japan:', max_genre_jp)
print('Other:', max_genre_other)
```

```python
# North America: Index(['Action', 'Sports', 'Shooter', 'Platform', 'Misc']
# Europe: Index(['Action', 'Sports', 'Shooter', 'Racing', 'Misc']
# Japan: Index(['Role-Playing', 'Action', 'Sports', 'Platform', 'Misc']
# Other: Index(['Action', 'Sports', 'Shooter', 'Racing', 'Misc']


# 각각의 TOP5 장르가 차지하는 비율 
# North America: 58.91 %
# Europe: 68.36 %
# Japan: 67.71 %
# Other: 72.16 %


```

```python
# 각 지역에서 top5 장르가 차지하는 비율 구하기 

# 각 지역의 전체 판매량 구하기
total_sales_na = df['NA_Sales'].sum()
total_sales_eu = df['EU_Sales'].sum()
total_sales_jp = df['JP_Sales'].sum()
total_sales_other = df['Other_Sales'].sum()

# 각 지역에서 선호 장르 top5의 판매량 구하기
sales_genre_na = df[df['Genre'].isin(max_genre_na)]['NA_Sales'].sum()
sales_genre_eu = df[df['Genre'].isin(max_genre_eu)]['EU_Sales'].sum()
sales_genre_jp = df[df['Genre'].isin(max_genre_jp)]['JP_Sales'].sum()
sales_genre_other = df[df['Genre'].isin(max_genre_other)]['Other_Sales'].sum()

# 각 지역에서 선호 장르 top5의 판매량이 전체 판매량에서 차지하는 비율 구하기
ratio_genre_na = sales_genre_na / total_sales_na
ratio_genre_eu = sales_genre_eu / total_sales_eu
ratio_genre_jp = sales_genre_jp / total_sales_jp
ratio_genre_other = sales_genre_other / total_sales_other

print('North America:', round(ratio_genre_na * 100, 2), '%')
print('Europe:', round(ratio_genre_eu * 100, 2), '%')
print('Japan:', round(ratio_genre_jp * 100, 2), '%')
print('Other:', round(ratio_genre_other * 100, 2), '%')
```

```python

df_sales = df[df['Year'].between(1995, 2016)]

df_grouped_sales = df_sales.groupby(['Year'])['Global_Sales'].sum().reset_index()
df_grouped_genre = df_sales.groupby(['Year', 'Genre'])['Global_Sales'].sum().reset_index()

df_merged = pd.merge(df_grouped_genre, df_grouped_sales, on='Year')
df_merged['Sales_Ratio'] = df_merged['Global_Sales_x'] / df_merged['Global_Sales_y']

df_pivot = df_merged.pivot(index='Year', columns='Genre', values='Sales_Ratio')

# Define colors for each genre
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2',
          '#7f7f7f', '#bcbd22', '#17becf', '#b94650', '#cbe554', '#4fc4aa']

# Stacked bar plot
fig, ax = plt.subplots(figsize=(12, 8))
df_pivot.plot(kind='bar', color=colors, stacked=True, ax=ax)
ax.set_title('Genre Sales Ratio by Year (1995-2016)')
ax.set_xlabel('Year')
ax.set_ylabel('Genre Sales Ratio')
ax.legend(title='Genre', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.show()
```
